```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/2010-2011 Solar home electricity data(perCustomer).csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# GC : General Consumption
GC = df.iloc[:,140:440]

# GG : Gross Generation
GG = df.iloc[:,440:] 

# CL : Controlled Load consumption
CL = df.iloc[:,1:140]
```

```python
GC.head()
```

```python
GG.head()
```

```python
CL.head()
```

```python
# 1row = 30 minute
# 2 row = 1 hour
# 48 row = 1 day
# 336 row = 1 weeks
```

```python
# GC 3d plot
daterow = 48
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

for i in range(100):
    x = np.arange(daterow)
    z = np.ones(daterow) + i+6
    y = GC.iloc[0:daterow,i+6]
    ax.plot(x,z,y, alpha=0.7)
ax.set_xlabel('Time')
ax.set_ylabel('Customer')
ax.set_zlabel('General Consumption')

# Get rid of the panes
ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
plt.show()
```

```python
# GC 2d plot
daterow = 48
plt.figure(figsize=(12,4))
for i in range(100):
    plt.plot(GC.iloc[0:daterow,i], alpha=0.7)
plt.xlabel('Time')
plt.ylabel('General Consumption')
plt.show()
```
