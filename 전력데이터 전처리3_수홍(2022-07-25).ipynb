```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/2010-2011 Solar home electricity data(perCustomer).csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# GC : General Consumption
GC = df.iloc[:,140:440]

# GG : Gross Generation
GG = df.iloc[:,440:] 

# CL : Controlled Load consumption
CL = df.iloc[:,1:140]
```

```python
surplus = GG.values - GC.values
```

```python
surplus = pd.DataFrame(surplus)
```

```python
surplus_names = []
#colnames.append("Load(Customer" +  str(int(c_customer[j])) + ":"+ c_category[k] + ")")
for i in range(len(surplus.columns)) :
    surplus_names.append(f"Load(Customer{i+1}:surplus)")
```

```python
surplus.columns = surplus_names
print(surplus.columns)
```

```python
date = df['date'][0:17520]
preprocessed_data = pd.concat([date, surplus], axis=1)
```

```python
preprocessed_data.to_csv('2010-2011 Solar home electricity data(surplus).csv', index=False)
```
