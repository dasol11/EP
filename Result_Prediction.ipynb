```python
GC_rbem_mse = mean_squared_error(GC_result['RBEM'], GC_result['Y'])
GC_ridge_mse = mean_squared_error(GC_result['Ridge'], GC_result['Y'])
GC_rf_mse = mean_squared_error(GC_result['RF'], GC_result['Y'])
GC_svr_mse = mean_squared_error(GC_result['SVR'], GC_result['Y'])
GC_arima_mse = mean_squared_error(GC_result['ARIMA'], GC_result['Y'])
GC_mean_mse = (GC_ridge_mse + GC_rf_mse + GC_svr_mse + GC_arima_mse) / 4
GC_lstm_mse = mean_squared_error(GC_result['LSTM'], GC_result['Y'])
GC_lstm_mse = mean_squared_error(GC_result['LSTM'][0:15000], GC_result['Y'][0:15000])

print('GC_RBEM MSE : %3f' %GC_rbem_mse)
print('GC_Ridge MSE : %3f' %GC_ridge_mse)
print('GC_RF MSE : %3f' %GC_rf_mse)
print('GC_SVR MSE : %3f' %GC_svr_mse)
print('GC_ARIMA MSE : %3f' %GC_arima_mse)
print('GC_Mean MSE : %3f' %GC_mean_mse)
print('GC_LSTM MSE : %3f' %GC_lstm_mse)
```

```python
GG_rbem_mse = mean_squared_error(GG_result['RBEM'], GG_result['Y'])
GG_ridge_mse = mean_squared_error(GG_result['Ridge'], GG_result['Y'])
GG_rf_mse = mean_squared_error(GG_result['RF'], GG_result['Y'])
GG_svr_mse = mean_squared_error(GG_result['SVR'], GG_result['Y'])
GG_arima_mse = mean_squared_error(GG_result['ARIMA'], GG_result['Y'])
GG_mean_mse = (GG_ridge_mse + GG_rf_mse + GG_svr_mse + GG_arima_mse) / 4
GG_lstm_mse = mean_squared_error(GG_result['LSTM'], GG_result['Y'])

print('GG_RBEM MSE : %3f' %GG_rbem_mse)
print('GG_Ridge MSE : %3f' %GG_ridge_mse)
print('GG_RF MSE : %3f' %GG_rf_mse)
print('GG_SVR MSE : %3f' %GG_svr_mse)
print('GG_ARIMA MSE : %3f' %GG_arima_mse)
print('GG_Mean MSE : %3f' %GG_mean_mse)
print('GG_LSTM MSE : %3f' %GG_lstm_mse)
```
